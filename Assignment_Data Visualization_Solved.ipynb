```python
# importing necessary libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
# reading data from local
df = pd.read_csv('F:/iNeuron DS/Assignments/dataSet/datasets_19_420_Iris.csv')
```

```python
# data view
df.head()
```

```python
# data count, balanced dataset
df.count()
```

```python
iris = df.drop('Id', axis=1)
```

```python
iris.head(2)
```

```python
# Separating features column and target into ndarray

X = iris.iloc[:, :-1].values
y = iris.iloc[:,4].values
```

```python
# standardize the data (mean =0 and variance = 1) before PCA applied
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# label encoding for target column - converting each values in the coulmn into numbers
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder()
y = en.fit_transform(y)
```

```python
# transform data dimension to 3D, now applying PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_reduced = pca.fit_transform(X)
```

```python
fig = plt.figure(1, figsize =(7,6))
ax = Axes3D(fig, elev =40, azim =-130)
ax.scatter(X_reduced[:,0], X_reduced[:,1], X_reduced[:,2], c=y, cmap = plt.cm.rainbow_r)
ax.set_title('3D PCA representation')
ax.set_xlabel('1st eigen vector')
ax.set_ylabel('2nd eigen vector')
ax.set_zlabel('3rd eigen vector')
plt.show()
```
