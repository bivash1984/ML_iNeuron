### Scipy: 
#### We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below. 
### Task: 
#### 1. fitting it to the periodic function 
#### 2. plot the fit 
#### Data 
#### Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25 
#### Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# maximum and minimum temperatures are taken into numpy array
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
months = np.arange(12) # this  arange function will create 12 numbers
plt.plot(months, max_temp, 'go')
plt.plot(months, min_temp, 'bo')
plt.xlabel('Months', fontsize = 15)
plt.ylabel('Temperature (C)', fontsize = 15)
```

```python
from scipy import optimize
def model(times, avg, ampl, time_offset): # creating model function for the cos curve
    return (ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()) + avg)

opt_max, cov_max = optimize.curve_fit(model, months, max_temp, [40, 20, 0]) # optimal values and covariance of optimal value
opt_min, cov_min = optimize.curve_fit(model, months, min_temp, [-40, 20, 0]) # optimal values and covariance of optimal value
```

```python
# plotting best fit line / curve

days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, max_temp, 'go')
plt.plot(days, model(days, *opt_max), 'm-')
plt.plot(months, min_temp, 'bo')
plt.plot(days, model(days, *opt_min), 'y-')
plt.xlabel('Months', fontsize = 15)
plt.ylabel('Temperature (C)', fontsize = 15)

plt.show()
```

### 1. Create a pie chart presenting the male/female proportion 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
df = titanic['sex'].value_counts()
df
```

```python
df.plot.pie(subplots = True, figsize = (5,5), fontsize = 15, autopct = '%.2f%%')
plt.title('Presenting male/female proportion')
```

### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender 

```python
titanic=titanic.dropna(subset=['sex'])
colors = {'male' : 'blue', 'female' : 'red'}
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, c=titanic['sex'].map(colors))
plt.xlabel('age', fontsize=15)
plt.ylabel('fare', fontsize=15)
plt.title('Fare paid and the Age', fontsize=20)


#plt.legend (['female', 'male'])
```
